```python
import pandas as pd
fruit = pd.DataFrame(
    {'Num' : [123,456,789,1011,1112],
    'Fruit' : ['apple', 'banana', 'cherry', 'lemon', 'peach']})

grade = pd.DataFrame(
    {'Num' : [123,789,1314],
    'Grade' : ['A', 'B', 'c']})

price = pd.DataFrame(
    {'No' : [123,789,1314],
    'Grade' : [100, 200, 500]})
```

```python
df = pd.concat([fruit, grade])
df
```

```python
df = pd.merge(fruit, grade, on = 'Num', how = 'outer')
df
```

```python
df.isnull().sum()
```

```python
df = pd.merge(fruit, price, left_on = 'Num', right_on = 'No', how = 'outer')
df
```
